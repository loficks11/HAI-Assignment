```python
import numpy as np
import pandas as pd

data = {
    '날짜': ['3/20', '3/21', '3/22', '3/23', '3/24'],
    '매출 편차': np.random.randn(5),
    '손님 평균 연령': np.random.randint(10, 60, 5)
}
df = pd.DataFrame(data)
df
```

```python
df['매출 편차']
```

```python
df[df['손님 평균 연령'] >= 20]
```

```python
df.loc[2, '매출 편차'] = np.nan
df.iloc[4, 2] = np.nan
df
```

```python
df.isnull()
```

```python
df.fillna({'매출 편차': 0}, inplace = True)
df
```

```python
df.fillna({'손님 평균 연령': df['손님 평균 연령'].mean()}, inplace=True)
df
```
